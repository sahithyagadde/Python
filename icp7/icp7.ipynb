```
import soup as soup
from bs4 import BeautifulSoup
import os
import requests
#1
url="https://en.wikipedia.org/wiki/Google"
scode= requests.get(url)
ptext= scode.text
soup= BeautifulSoup(ptext,"html.parser")
text = soup.find_all(text=True)
output = ''
blacklist = [
	'[document]',
	'noscript',
	'header',
	'html',
	'meta',
	'head', 
	'input',
	'script',
	# there may be more elements you don't want, such as "style", etc.
]

for t in text:
	if t.parent.name not in blacklist:
		output += '{} '.format(t)

#2

f = open("outputfile1.txt","w")
f.write(str(soup.text))
f.close()
#3.a
print('Tokenization Output is')
stoken=nltk.sent_tokenize(output)
wtoken=nltk.word_tokenize(output)
for s in stoken:
  print(s)
for w in wtoken:
  print(w)
#3.b
print('Stemming Output is')
tagged=nltk.pos_tag(output)
print('POS Output is',tagged)
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

ps = PorterStemmer()
sb = SnowballStemmer('english')
lc = LancasterStemmer()
for x in wtoken:
    print('PorterStemmer',ps.stem(x))
    print('Snowball Stemmer',sb.stem(x))
    print('Lancaster Stemmer',lc.stem(x))
#3.c
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
print('lemmatization Output is')
wtokens = nltk.word_tokenize(output)

lemmatizer = WordNetLemmatizer()

for x in wtokens:
    print(lemmatizer.lemmatize(x))

#3.d
nltk.download('maxent_ne_chunker')
nltk.download('words')
print('Trigram output is')
stokens = nltk.sent_tokenize(output)

for sent in stokens:
    wtokens = nltk.word_tokenize(sent)
    for j in range(len(wtokens) - 2):
        print(wtokens[j], wtokens[j + 1], wtokens[j + 2])

#3.e
from nltk import wordpunct_tokenize, pos_tag, ne_chunk
print('Named Entity Recognition Output is:')
stokens = nltk.sent_tokenize(output)

for i in stokens:
    print(ne_chunk(pos_tag(wordpunct_tokenize(i))))
```

```
from sklearn.neighbors import KNeighborsClassifier
from nltk.corpus import stopwords
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB

twenty_train = fetch_20newsgroups(subset='train', shuffle=True)

tfidf_Vect = TfidfVectorizer()
X_train_tfidf = tfidf_Vect.fit_transform(twenty_train.data)
clf = KNeighborsClassifier()
clf.fit(X_train_tfidf, twenty_train.target)

twenty_test = fetch_20newsgroups(subset='test', shuffle=True)
X_test_tfidf = tfidf_Vect.transform(twenty_test.data)

predicted = clf.predict(X_test_tfidf)

score = metrics.accuracy_score(twenty_test.target, predicted)
print(score)



#4.b
twenty_train = fetch_20newsgroups(subset='train', shuffle=True)

tfidf_Vect = TfidfVectorizer(ngram_range=(1,2))

X_train_tfidf = tfidf_Vect.fit_transform(twenty_train.data)

clf = MultinomialNB()
clf.fit(X_train_tfidf, twenty_train.target)

twenty_test = fetch_20newsgroups(subset='test', shuffle=True)
X_test_tfidf = tfidf_Vect.transform(twenty_test.data)

predicted = clf.predict(X_test_tfidf)

score = metrics.accuracy_score(twenty_test.target, predicted)
print(score)

#4.c
twenty_train = fetch_20newsgroups(subset='train', shuffle=True)

tfidf_Vect = TfidfVectorizer(stop_words='english')
X_train_tfidf = tfidf_Vect.fit_transform(twenty_train.data)

clf = MultinomialNB()
clf.fit(X_train_tfidf, twenty_train.target)

twenty_test = fetch_20newsgroups(subset='test', shuffle=True)
X_test_tfidf = tfidf_Vect.transform(twenty_test.data)

predicted = clf.predict(X_test_tfidf)

score = metrics.accuracy_score(twenty_test.target, predicted)
print(score)
```
